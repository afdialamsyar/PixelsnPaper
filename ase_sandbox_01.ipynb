```python
Ni211_surf = ase.build.fcc211("Ni", (6, 6, 4), vacuum=4.0)
# Ini akan memodifikasi Ni111_surf
cell_x = Ni111_surf.cell[0][0]
cell_y = Ni111_surf.cell[1][1]
pos_center = (0.5*cell_x-0.1, 0.5*cell_y+3)
print(f"cell_x = {cell_x}, cell_y = {cell_y}")
print(f"pos_center = {pos_center}")
ase.build.add_adsorbate(
    Ni211_surf, CH3OH_mol, 2.0, position=pos_center, mol_index=0)
view(Ni211_surf, viewer="x3d")
```
